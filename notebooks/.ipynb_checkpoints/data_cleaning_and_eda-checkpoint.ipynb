```python
def column_reformatting(df):
    cols = df.columns.tolist()
    new_cols = []
    for col in cols:
        new_cols.append(re.sub(r'(?<!^)(?=[A-Z])', '_', col).lower())
    df.columns = new_cols
    return df.head
```

```python
def one_hot_encoding(df, categorical_feature_list):
    for cat in categorical_feature_list:
            dummies = pd.get_dummies(df[[cat]])
            df = pd.concat([df, dummies], axis=1)
    return df
```

```python
df = one_hot_encoding(df, categorical_features)
```

```python
def one_hot_encoding(df, categorical_feature):
    dummies = pd.get_dummies(df[[categorical_feature]])
    new_df = pd.concat([df, dummies], axis=1)
    return new_df
```
