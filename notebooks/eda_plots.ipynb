```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
```

```python
plt.rc("font", size=16)
plt.style.use('ggplot')
```

```python
df = pd.read_csv("./data/employee_attrition.csv")

disc_col = []
for column in df.columns:
    if df[column].dtypes != object and df[column].nunique() < 30:
        print(f"{column} : {df[column].unique()}")
        disc_col.append(column)
        print("====================================")
disc_col.remove('Attrition')

plt.figure(figsize=(20, 20))
```

```python
for i, column in enumerate(disc_col, 1):
    plt.subplot(4, 4, i)
    df[df["Attrition"] == 'No'][column].hist(bins=35, color='blue', label='Attrition = NO', alpha=0.6)
    df[df["Attrition"] == 'Yes'][column].hist(bins=35, color='red', label='Attrition = YES', alpha=0.6)
    plt.legend()
    plt.xlabel(column)
```

```python
cont_col = []
for column in df.columns:
    if df[column].dtypes != object and df[column].nunique() > 30:
        print(f"{column} : Minimum: {df[column].min()}, Maximum: {df[column].max()}")
        cont_col.append(column)
        print("====================================")

plt.figure(figsize=(20, 10))

for i, column in enumerate(cont_col, 1):
    plt.subplot(2, 4, i)
    df[df["Attrition"] == 'No'][column].hist(bins=35, color='blue', label='Attrition = NO', alpha=0.6)
    df[df["Attrition"] == 'Yes'][column].hist(bins=35, color='red', label='Attrition = YES', alpha=0.6)
    plt.legend()
    plt.xlabel(column)

```

```python
import pandas as pd
df = pd.read_csv("./data/employee_attrition.csv")
```

```python
total_records = len(df)
```

```python
columns = ["Age","MaritalStatus","WorkLifeBalance","DistanceFromHome","MonthlyIncome",
           "YearsAtCompany","BusinessTravel","Department"]
```

```python

sns.countplot(x='BusinessTravel', hue='Attrition', data=df)
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(20, 10))
# plt.rcParams.update({"font", size=22})
plt.style.use('ggplot')

for i, column in enumerate(columns, 1):
    plt.subplot(2, 4, i)
    df[df["Attrition"] == 'No'][column].hist(color='darkblue', label='Attrition = No',alpha=0.7)
    df[df["Attrition"] == 'Yes'][column].hist(color='darkred', label='Attrition = Yes', alpha=0.8)
    plt.legend()
    plt.xlabel(column, fontsize=16)
    plt.xticks(fontsize=12)
    plt.ylabel("Employee Count", fontsize=16)
    plt.tight_layout()
```
