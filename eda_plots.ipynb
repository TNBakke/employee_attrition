```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("./data/employee_attrition.csv")
```

```python
plt.rc("font", size=16)
```

```python
import seaborn as sns
```

```python
df['Attrition'].value_counts()
```

```python
sns.countplot(x='Attrition', data=df)
plt.show()
```

```python
sns.set(style="white")
sns.set(style="whitegrid", color_codes=True)
```

```python
sns.countplot(x='Attrition', data=df)
plt.show()
```

```python
%matplotlib inline
pd.crosstab(df.Age,df.Attrition).plot(kind='bar', figsize=(15,10))
plt.title('Attrition by Age')
plt.xlabel('Age')
plt.ylabel('Employee Count')
plt.show()
```

```python
%matplotlib inline
pd.crosstab(df.DistanceFromHome,df.Attrition).plot(kind='bar', figsize=(15,10))
plt.title('Attrition by Commute Length')
plt.xlabel('Commute Length (Miles)')
plt.ylabel('Employee Count')
plt.show()
```

```python
%matplotlib inline
pd.crosstab(df.Department,df.Attrition).plot(kind='bar', figsize=(15,10))
plt.title('Attrition by Department')
plt.xlabel('Department')
plt.ylabel('Employee Count')
plt.xticks(rotation=45)
plt.show()
```

```python
%matplotlib inline
pd.crosstab(df.OverTime,df.Attrition).plot(kind='bar', figsize=(15,10))
plt.title('Attrition by Overtime Work')
plt.xlabel('Overtime')
plt.ylabel('Employee Count')
plt.xticks(rotation=45)
plt.show()
```

```python
%matplotlib inline
pd.crosstab(df.WorkLifeBalance,df.Attrition).plot(kind='bar', figsize=(15,10))
plt.title('Attrition by Work Life Balance')
plt.xlabel('Work Life Balance (4 is Best)')
plt.ylabel('Employee Count')
plt.xticks(rotation=45)
plt.show()
```

```python
%matplotlib inline
pd.crosstab(df.JobSatisfaction,df.Attrition).plot(kind='bar', figsize=(15,10))
plt.title('Attrition by Job Satisfaction')
plt.xlabel('Job Satisfaction (4 is Best)')
plt.ylabel('Employee Count')
plt.xticks(rotation=45)
plt.show()
```

```python
%matplotlib inline
pd.crosstab(df.Gender,df.Attrition).plot(kind='bar', figsize=(15,10))
plt.title('Attrition by Gender')
plt.xlabel('Gender')
plt.ylabel('Employee Count')
plt.xticks(rotation=45)
plt.show()
```
